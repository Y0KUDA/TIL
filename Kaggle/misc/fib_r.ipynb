```r
fib <- function(n)ifelse(n<=1,1,fib(n-2)+fib(n-1))

fib_t <- function(n){
    start <- proc.time()
    fib(n)
    finish <- proc.time()
    return ((finish - start)[1])
}
```

```r
score <- read.csv("./fib_score.csv",header=T)
```

```r
t=fib_t(10)
```

```r
for(i in 1:30){
    score$R[i]=fib_t(i)
}
```

```r
write.csv(score,"./fib_score.csv")
```
